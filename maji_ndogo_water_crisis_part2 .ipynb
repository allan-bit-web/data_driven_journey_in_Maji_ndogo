using the employee table to count how many employees live in each town.

```python
%%sql

SELECT DISTINCT
   town_name,
   
   COUNT(employee_name) OVER (PARTITION BY town_name ) As num_employees
FROM
      md_water_services.employee;
      
```

```python
%%sql

SELECT DISTINCT
   town_name,
   
   COUNT(employee_name) As num_employees
FROM
      md_water_services.employee
GROUP BY town_name;
      
```

```python
%%sql

SELECT DISTINCT
    assigned_employee_id,
    COUNT(visit_count) OVER (PARTITION BY  assigned_employee_id ) AS number_of_visits
FROM
     visits 
ORDER BY number_of_visits DESC
LIMIT 3;
```

```python
%%sql

SELECT DISTINCT
    assigned_employee_id,
    COUNT(visit_count) AS number_of_visits
FROM
     visits 
GROUP BY assigned_employee_id
ORDER BY number_of_visits DESC
LIMIT 3;
```

Query that looks up the employee's info:names,email addresses and phone numbers for our top 3 employee_ids with the highest number of locations visited

```python
%%sql

SELECT DISTINCT
     *
FROM
     employee
LIMIT 2;
```

```python
%%sql

SELECT DISTINCT
    v.assigned_employee_id AS employee_id ,
     e.employee_name,
     e.email AS email,
     e.phone_number AS phone_number,
    COUNT(visit_count) AS number_of_visits
FROM
     visits v,employee e
WHERE v.assigned_employee_id = e.assigned_employee_id 
GROUP BY employee_id
ORDER BY number_of_visits DESC
LIMIT 3;
```
