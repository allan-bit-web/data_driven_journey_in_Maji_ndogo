```python
%%sql

SELECT DISTINCT
    assigned_employee_id,
    COUNT(visit_count) OVER (PARTITION BY  assigned_employee_id ) AS number_of_visits
FROM
     visits 
ORDER BY number_of_visits DESC
LIMIT 3;
```

```python
%%sql

SELECT DISTINCT
    assigned_employee_id,
    COUNT(visit_count) AS number_of_visits
FROM
     visits 
GROUP BY assigned_employee_id
ORDER BY number_of_visits DESC
LIMIT 3;
```

Query that looks up the employee's info:names,email addresses and phone numbers for our top 3 employee_ids with the highest number of locations visited

```python
%%sql

SELECT DISTINCT
     *
FROM
     employee
LIMIT 2;
```

```python
%%sql

SELECT DISTINCT
    v.assigned_employee_id AS employee_id ,
     e.employee_name,
     e.email AS email,
     e.phone_number AS phone_number,
    COUNT(visit_count) AS number_of_visits
FROM
     visits v,employee e
WHERE v.assigned_employee_id = e.assigned_employee_id 
GROUP BY employee_id
ORDER BY number_of_visits DESC
LIMIT 3;
```

The way I look at this table; we have access to different water source types and the number of people using each source.
These are the questions that I am curious about.
1. How many people did we survey in total?
2. How many wells, taps and rivers are there?
3. How many people share particular types of water sources on average?
4. How many people are getting water from each type of source?

```python
%%sql

SELECT 
     SUM(number_of_people_served) AS Total_number_served
FROM 
     water_source
LIMIT 5;
```

```python
%%sql

SELECT DISTINCT 
  type_of_water_source,
  COUNT(type_of_water_source) AS number_of_sources
FROM 
     water_source
GROUP BY type_of_water_source
LIMIT 5;
```

Average number of people that are served by each water source

```python
%%sql

SELECT DISTINCT 
  type_of_water_source,
  ROUND(AVG(number_of_people_served)) AS ave_people_per_source
FROM 
     water_source
GROUP BY type_of_water_source;
```

Calculating the total number of people served by each type of water source in total and ordering in descending order

```python
%%sql

SELECT DISTINCT 
  type_of_water_source,
  SUM(number_of_people_served) AS population_served
FROM 
     water_source
GROUP BY type_of_water_source
ORDER BY population_served DESC;
```

To make our insights easier to understand we convert population_served into percentages

```python
%%sql

SELECT DISTINCT 
  type_of_water_source,
  ROUND((SUM(number_of_people_served)/27628140)*100) AS percentage_people_per_source
FROM 
     water_source
GROUP BY type_of_water_source
ORDER BY percentage_people_per_source DESC;
```
